```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.rcParams["font.size"] = 42
```

```python
def generate_box_plot(encryption):
    fr_df = pd.read_csv(f"../ssim_value/ssim_fr_attack_{encryption}.csv")
    gan_df = pd.read_csv(f"../ssim_value/ssim_gan_attack_{encryption}.csv")
    itn_df = pd.read_csv(f"../ssim_value/ssim_itn_attack_{encryption}.csv")
    
    ssim_df = pd.concat({"FR-Attack": fr_df["ssim"],
                         "GAN-Attack": gan_df["ssim"],
                         "ITN-Attack": itn_df["ssim"]}, axis=1)
    ssim_df = ssim_df.abs()
    ssim_df = pd.melt(ssim_df)
    fig, ax = plt.subplots(figsize=(26, 15))
    sns.boxplot(x="variable", y="value", data=ssim_df, ax=ax, linewidth=3);
    ax.set_xlabel("");
    ax.set_ylabel("SSIM");
```

# LE

```python
encryption = "le"
```

```python
generate_box_plot(encryption)
```

# ELE

```python
encryption = "ele"
```

```python
generate_box_plot(encryption)
```
